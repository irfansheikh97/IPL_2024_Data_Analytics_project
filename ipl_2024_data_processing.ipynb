#### (1) Match Results Data processing

```python
# important necessary libraries
import pandas as pd
import json
```

```python
with open('IPL2024_DATASET/ipl_2024_match_results.json', 'r') as f:
	data = json.load(f)
match_df = pd.DataFrame(data['matchSummary'])
match_df
```

```python
match_df.shape
```

```python
match_df.rename({'scorecard':'match_id'}, axis=1, inplace=True)
```

```python
match_df = match_df.iloc[::-1].reset_index(drop=True)
```

```python
match_df['match_id'] = range(1, len(match_df) + 1)
```

```python
# Initialize a dictionary to keep track of match IDs
match_id_dict = {}
match_id_counter = 1

# Function to generate match ID
def get_match_id(row):
    global match_id_counter
    match_tuple = tuple(sorted([row['team1'], row['team2']]))
    if match_tuple not in match_id_dict:
        match_id_dict[match_tuple] = match_id_counter
        match_id_counter += 1
    return match_id_dict[match_tuple]

# Apply the function to create the match_id column
match_df['match_id'] = match_df.apply(get_match_id, axis=1)

# team name change from short to full
team_name_mappings = {
    'MI': 'Mumbai Indians',
    'CSK': 'Chennai Super Kings',
    'RCB': 'Royal Challengers Bengaluru',
    'KKR': 'Kolkata Knight Riders',
	'SRH': 'Sunrisers Hyderabad',
	'RR': 'Rajasthan Royals',
	'DC': 'Delhi Capitals',
	'LSG': 'Lucknow Super Giants',
	'GT' : 'Gujarat Titans',
	'Punjab Kings': 'Punjab Kings'
}

match_df['team1'] = match_df['team1'].map(team_name_mappings)
match_df['team2'] = match_df['team2'].map(team_name_mappings)
match_df['winner'] = match_df['winner'].map(team_name_mappings)
match_df
```

```python
match_ids_dict = {}

for index, row in match_df.iterrows():
	key1 = row['team1'] + ' Vs ' + row['team2']
	key2 = row['team2'] + ' Vs ' + row['team1']
	# match_ids_dict[row['team1'] + " Vs " + row['team2']] = row['match_id']
	if key1 not in match_ids_dict.keys():
		match_ids_dict[key1] = row['match_id']
		
	if key2 not in match_ids_dict.keys():
		match_ids_dict[key2] = row['match_id']
		
match_ids_dict
```

```python
match_df.to_csv('IPL2024_CSV_FILES/match_summary.csv', index = False)
```
